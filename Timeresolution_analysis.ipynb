```python
sigma_t  = np.array([])
sigma_t_error = np.array([])
centers = np.array([])
xerrors = np.array([])


photobins = np.geomspace(40,100,7)
#print(photobins)

for i in range(len(photobins)-1):
    edges = (photobins[i],photobins[i+1])
    
    
    
    #print(edges)
    #df.query('ch3_P2P_smooth_PE>%s'%(limit))['time_diff'])
    
    hist, bin_edges = np.histogram(df.query('signal_strength>%2.2f and signal_strength<%2.2f'%(edges[0],edges[1]))['time_diff'],bins=10,range=(-2,2))
    #hist, bin_edges = np.histogram(df.query('signal_strength>%2.2f and signal_strength<%2.2f and ch3_P2P_smooth_PE>%s'%(edges[0],edges[1],limit))['time_diff'],bins=10,range=(-2,2))
    
    bin_centers = (bin_edges[:-1] + bin_edges[1:])/2
    p0 = [1,0,.5]
    coeff, var_matrix = curve_fit(gauss, bin_centers, hist, p0=p0)
                            #(start,end,line quality keep at 100)
    hist_fit = gauss(np.linspace(-2, 2, 100), *coeff)

    sigma_t = np.append(sigma_t, 1000.0*abs(coeff[2])/np.sqrt(2)) #divide by sqrt(2) to transform to single detector resolution?
    sigma_t_error = np.append(sigma_t_error, 1000.0*np.sqrt(var_matrix[2][2])/np.sqrt(2))
    centers = np.append(centers, (edges[0]+edges[1])/2.0)
    xerrors = np.append(xerrors, (edges[1]-edges[0])/2.0)



def resolution(x, *p):
    A, l = p
    return A/np.sqrt(x)
    
fig = plt.figure(figsize=(8,6))
plt.errorbar(centers,y=sigma_t,xerr=xerrors,yerr=sigma_t_error,fmt='o')
coeff, var_matrix = curve_fit(resolution, centers, sigma_t, p0=[100,5],sigma=sigma_t_error)

#exeptionally


#print(np.linspace(45,70))
#print(coeff)
#print(expo(np.linspace(40,70,10),*coeff))

plt.plot(photobins, resolution(photobins, *coeff),label='A/$\sqrt(E)$ fit')


plt.ylim([0,400])
plt.xlabel('Signal strength (Photoelectrons)')
plt.ylabel('Time resolution [ps]')
#plt.legend()
#plt.savefig('Nov18Sourcetimestrength2.png')
```

```python
#SINGAL EFF INDIVIDUAL

from scipy.optimize import curve_fit
import numpy as np

hist, bin_edges = np.histogram(df.query('signal_strength>60 and signal_strength<70')['time_diff'],bins=10,range=(-2,2))


#Cosmic
#hist, bin_edges = np.histogram(df['ch3_P2P_smooth_PE'],bins=10,range=(0,10))

bin_centers = (bin_edges[:-1] + bin_edges[1:])/2

def gauss(x, *p):
    A, mu, sigma = p
    return A*np.exp(-(x-mu)**2/(2.*sigma**2))

p0 = [1,0,.5]


coeff, var_matrix = curve_fit(gauss, bin_centers, hist, p0=p0)
                            #(start,end,line quality keep at 100)
hist_fit = gauss(np.linspace(-2, 2, 100), *coeff)
plt.errorbar(bin_centers, y=hist, yerr=np.sqrt(hist),fmt='o')
plt.plot(np.linspace(-2,2,100), hist_fit,)
plt.xlabel("Photoelectrons")
plt.ylabel("Count")



plt.legend(['$\mu$=%2.0f ps +/- %2.0f ps, \n$\sigma$= %2.0f ps +/- %2.0f ps '%(1000*coeff[1],
                                                                               1000*np.sqrt(var_matrix[1][1]),
                                                                               1000*np.abs(coeff[2]),
                                                                               1000*np.sqrt(var_matrix[2][2]))],
           loc=2,prop={'size': 18})       

#plt.title('Signal Strength eff for Back-to-back Cosmic Run ')
plt.xlabel('Time Difference [ns]')
plt.ylabel('counts')

sigma1=(1000*coeff[2])
print(sigma1)

#hist, bin_edges = np.histogram(df.query('signal_strength>10')['time_diff'],bins=10,range=(-2,2))
#hist, bin_edges = np.histogram(df.query('signal_strength>40 and signal_strength<50')['time_diff'],bins=10,range=(-2,2))
hist, bin_edges = np.histogram(df.query('signal_strength>50 and signal_strength<60')['time_diff'],bins=10,range=(-2,2))



#Cosmic
#hist, bin_edges = np.histogram(df['ch3_P2P_smooth_PE'],bins=10,range=(0,10))

bin_centers = (bin_edges[:-1] + bin_edges[1:])/2

def gauss(x, *p):
    A, mu, sigma = p
    return A*np.exp(-(x-mu)**2/(2.*sigma**2))

p0 = [1,0,.5]


coeff, var_matrix = curve_fit(gauss, bin_centers, hist, p0=p0)
                            #(start,end,line quality keep at 100)
hist_fit = gauss(np.linspace(-2, 2, 100), *coeff)
plt.errorbar(bin_centers, y=hist, yerr=np.sqrt(hist),fmt='o')
plt.plot(np.linspace(-2,2,100), hist_fit,)
plt.xlabel("Photoelectrons")
plt.ylabel("Count")



plt.legend(['$\mu$=%2.0f ps +/- %2.0f ps, \n$\sigma$= %2.0f ps +/- %2.0f ps '%(1000*coeff[1],
                                                                               1000*np.sqrt(var_matrix[1][1]),
                                                                               1000*np.abs(coeff[2]),
                                                                               1000*np.sqrt(var_matrix[2][2]))],
           loc=2,prop={'size': 18})       

#plt.title('Signal Strength eff for Back-to-back Cosmic Run ')
plt.xlabel('Time Difference [ns]')
plt.ylabel('counts')

sigma2=(1000*coeff[2])
print(sigma2)

#hist, bin_edges = np.histogram(df.query('signal_strength>10')['time_diff'],bins=10,range=(-2,2))
hist, bin_edges = np.histogram(df.query('signal_strength>40 and signal_strength<50')['time_diff'],bins=10,range=(-2,2))



#Cosmic
#hist, bin_edges = np.histogram(df['ch3_P2P_smooth_PE'],bins=10,range=(0,10))

bin_centers = (bin_edges[:-1] + bin_edges[1:])/2

def gauss(x, *p):
    A, mu, sigma = p
    return A*np.exp(-(x-mu)**2/(2.*sigma**2))

p0 = [1,0,.5]


coeff, var_matrix = curve_fit(gauss, bin_centers, hist, p0=p0)
                            #(start,end,line quality keep at 100)
hist_fit = gauss(np.linspace(-2, 2, 100), *coeff)
plt.errorbar(bin_centers, y=hist, yerr=np.sqrt(hist),fmt='o')
plt.plot(np.linspace(-2,2,100), hist_fit,)
plt.xlabel("Photoelectrons")
plt.ylabel("Count")



plt.legend(['$\mu$=%2.0f ps +/- %2.0f ps, \n$\sigma$= %2.0f ps +/- %2.0f ps '%(1000*coeff[1],
                                                                               1000*np.sqrt(var_matrix[1][1]),
                                                                               1000*np.abs(coeff[2]),
                                                                               1000*np.sqrt(var_matrix[2][2]))],
           loc=2,prop={'size': 18})       

#plt.title('Signal Strength eff for Back-to-back Cosmic Run ')
plt.xlabel('Time Difference [ns]')
plt.ylabel('counts')

sigma3=(1000*coeff[2])
print(sigma3)
hist, bin_edges = np.histogram(df.query('signal_strength>10')['time_diff'],bins=10,range=(-2,2))



#Cosmic
#hist, bin_edges = np.histogram(df['ch3_P2P_smooth_PE'],bins=10,range=(0,10))

bin_centers = (bin_edges[:-1] + bin_edges[1:])/2

def gauss(x, *p):
    A, mu, sigma = p
    return A*np.exp(-(x-mu)**2/(2.*sigma**2))

p0 = [1,0,.5]


coeff, var_matrix = curve_fit(gauss, bin_centers, hist, p0=p0)
                            #(start,end,line quality keep at 100)
hist_fit = gauss(np.linspace(-2, 2, 100), *coeff)
plt.errorbar(bin_centers, y=hist, yerr=np.sqrt(hist),fmt='o')
plt.plot(np.linspace(-2,2,100), hist_fit,)
plt.xlabel("Photoelectrons")
plt.ylabel("Count")



plt.legend(['$\mu$=%2.0f ps +/- %2.0f ps, \n$\sigma$= %2.0f ps +/- %2.0f ps '%(1000*coeff[1],
                                                                               1000*np.sqrt(var_matrix[1][1]),
                                                                               1000*np.abs(coeff[2]),
                                                                               1000*np.sqrt(var_matrix[2][2]))],
           loc=2,prop={'size': 18})       

#plt.title('Signal Strength eff for Back-to-back Cosmic Run ')
plt.xlabel('Time Difference [ns]')
plt.ylabel('counts')

sigma4=(1000*coeff[2])
print(sigma4)
```
